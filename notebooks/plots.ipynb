# Dotplot

```python
lda = pd.read_csv('../dash/data/lda.csv')
lda.head()
```

```python
trace1 = {"x": lda.R, 
          "y": lda.words_lda, 
          "marker": {"color": "red", "size": 12}, 
          "mode": "markers", 
          "name": "Republican", 
          "type": "scatter"
}

trace2 = {"x": lda.D, 
          "y": lda.words_lda, 
          "marker": {"color": "blue", "size": 12}, 
          "mode": "markers", 
          "name": "Democrat", 
          "type": "scatter"
}

data = go.Data([trace1, trace2])
layout = {"title": "Sentiment Disparity by Topic and Party", 
          "xaxis": {"title": "Sentiment Polarity", }, 
          "yaxis": {"title": ""}}

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, fileopt='overwrite')
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

topics = lda.words_lda
r = lda.R
d = lda.D

trace0 = go.Scatter(
    x=r,
    y=topics,
    mode='markers',
    name='Republican',
    marker=dict(
        color='red',
        line=dict(
            color='rgba(156, 165, 196, 1.0)',
            width=1,
        ),
        symbol='circle',
        size=10,
    )
)
trace1 = go.Scatter(
    x=d,
    y=topics,
    mode='markers',
    name='Democrat',
    marker=dict(
        color='blue',
        line=dict(
            color='rgba(217, 217, 217, 1.0)',
            width=1,
        ),
        symbol='circle',
        size=10,
    )
)

data = [trace0, trace1]
layout = go.Layout(
    title="Sentiment Polarity by Topic and Party",
    xaxis=dict(
        showgrid=False,
        showline=True,
        linecolor='rgb(102, 102, 102)',
        titlefont=dict(
            color='rgb(204, 204, 204)'
        ),
        tickfont=dict(
            color='rgb(102, 102, 102)',
        ),
        autotick=False,
        dtick=10,
        ticks='outside',
        tickcolor='rgb(102, 102, 102)',
    ),
    margin=dict(
        l=400,
        r=40,
        b=50,
        t=80
    ),
    legend=dict(
        font=dict(
            size=10,
        ),
        yanchor='middle',
        xanchor='right',
    ),
    width=800,
    height=600,
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, fileopt='overwrite')
```
